```python
import numpy as np
import pandas as pd
# Load the datasets
df = pd.read_csv("https://raw.githubusercontent.com/wongwara/Jobseeker_Baymax/main/salary_fulltime/df_salary_fulltime1.csv")
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
max_salary = df['max_salary'].max()
print(max_salary)
```

```python
df['max_salary'] = df['max_salary'].astype(int)
df['min_salary'] = df['min_salary'].astype(int)
```

```python
df['min_salary']
```

```python
# Define the salary ranges based on the minimum and maximum salaries
salary_ranges = [(18000, 60000), (60000, 80000), (80000, 90000), (90000, 100000), 
                 (100000, 110000), (110000, 120000), (120000, 130000), (130000,140000),
                (140000,160000),(160000,180000),(180000,float('inf'))]

# Create a new column in the DataFrame to represent the salary range section based on the minimum and maximum salaries
df['salary_section'] = pd.cut(df[['min_salary', 'max_salary']].mean(axis=1), 
                              bins=[r[0] for r in salary_ranges] + [salary_ranges[-1][1]])

# Print the number of job listings in each salary range section
print(df['salary_section'].value_counts())
```

```python
df = df[df['salary_section'].notnull()]
```
